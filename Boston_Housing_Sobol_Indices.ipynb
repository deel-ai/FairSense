# Sobol Indices - Boston Housing dataset

## 1) Loading Boston Housing dataset, and fitting a Random Forest model

```python
from datasets import Datasets
from models import fit_random_forest


# Load the Boston Housing dataset
(x_train, y_train), (x_test, y_test) = Datasets.load_Boston_Housing()

# Create and fit a Random Forest model on the dataset
random_forest = fit_random_forest(x_train, y_train, x_test, y_test)
```

## 2) Compute Sobol indices on the model-dataset

```python
from methods import SobolIndices


# Get the Boston Housing dataset features list
feature_names = Datasets.list_Boston_Housing_features()

# Compute the Sobol Indices on the dataset
all_indices, main_indices = SobolIndices.compute(x_test, random_forest, feature_names)
```

## 3) Sobol indices analysis

```python
# Visualize main Sobol indices with barcharts
SobolIndices.visualize_barcharts(main_indices, feature_names)
```

```python
# Visualize main Sobol indices with piecharts
SobolIndices.visualize_piecharts(main_indices, feature_names)
```
